```python
#problem 1 -(a) 每月總降雨量甜甜圈圖
import pandas as pd
import matplotlib.pyplot as plt

fp = pd.read_csv("466900-2022.csv")
monthday = [31,28,31,30,31,30,31,31,30,31,30,31] #紀錄每月有幾天
water = [] #用來存每月降雨量
s =0
index =0
for i in monthday:
    for j in range(i): #把這個月中每天的降雨量加起來
        s += fp["降水量(mm)"][index]
        index += 1
    water.append(s) #算完存起來
    s = 0
    
month = [i for i in range(1,13)]
plt.pie(water,  #畫圖的數據
        labels = month, #設定標籤(月份)
        radius=1.2, #調整半徑
        autopct='%.1f%%', #算百分比
         pctdistance=0.7, #與中心的距離
        textprops={'size':14}, #文字大小
        wedgeprops={'linewidth':3,'edgecolor':'w','width':1} #設定餅圖
       ) 
plt.show()
```

```python
#problem 1 -(b) 氣溫降雨散布圖
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fp = pd.read_csv("466900-2022.csv")
    
month = [i for i in range(1,366)]
monthday = [31,28,31,30,31,30,31,31,30,31,30,31] #紀錄每月有幾天
water = [] #用來存每月降雨量
s =0
index =0
for i in monthday:
    for j in range(i): #把這個月中每天的降雨量加起來
        s += fp["降水量(mm)"][index]
        index += 1
    water.append(s) #算完存起來
    s = 0
    
    
#將圖表分割成兩個區塊
Fig = plt.figure(figsize=(24,8))
fig1 = Fig.add_axes([0,0,0.4,0.5])
fig2 = Fig.add_axes([0.4,0,0.4,0.5])

fig1.set_title("Annual precipitation 2565.5", fontdict={'fontsize': 20}) #標題

fig1.scatter(month,             # x軸資料
            fp["氣溫(℃)"],     # y軸資料
            c = fp["氣溫(℃)"],  # 做colorbar要傳的
             cmap = "Reds" #點顏色
            )

tick = fig1.set_xticks([60,120,180,240,300,360]) #設置刻度
labels = fig1.set_xticklabels([2,4,6,8,10,12], rotation=0, fontsize =8) #設置刻度值
colorbar1 = Fig.colorbar(fig1.collections[0], ax =fig1)  # 加入 colorbar

fig2.set_title("Precipitation", fontdict={'fontsize': 20}) #標題

fig2.scatter(month,             # x軸資料
            fp["降水量(mm)"],     # y軸資料
            c = fp["降水量(mm)"] ,     # 做colorbar要傳的
             cmap = "Blues"  #點顏色
            )

tick = fig2.set_xticks([60,120,180,240,300,360]) #設置刻度
labels = fig2.set_xticklabels([2,4,6,8,10,12], rotation=0, fontsize =8) #設置刻度值
colorbar2 = Fig.colorbar(fig2.collections[0] , ax = fig2)

#set label
fig1.set_xlabel("Month")
fig1.set_ylabel("Temperature (°C)")
fig2.set_xlabel("Month")
fig2.set_ylabel("Precipitation (mm)")
```

```python
#problem 1 -(c) ⽉均溫折線圖和⽉總降⽔量的長條圖
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fp = pd.read_csv("466900-2022.csv")

month = [i for i in range(1,13)]
monthday = [31,28,31,30,31,30,31,31,30,31,30,31] #紀錄每月有幾天
water = [] #用來存每月降雨量
tem = [] #算月均溫
s =0
s2 =0
index =0
for i in monthday:
    for j in range(i): #把這個月中每天的降雨量加起來
        s += fp["降水量(mm)"][index]
        s2 += fp["氣溫(℃)"][index]
        index += 1
    water.append(s) #算完存起來
    tem.append(s2/i) #算月均溫再存起來 
    s = 0
    s2 = 0

fig, ax1 = plt.subplots()
ax1.set_xlabel('Month')

ax1.bar(month,water,color ='b')
ax1.set_ylabel('tmonthly precipitation(mm)',color='b') #畫出y軸標籤
ax1.tick_params('y', colors='b') #畫出y軸顏色

ax2 = ax1.twinx() #同x軸 但不同y軸

ax2.plot(month,tem,'r-')
ax2.set_ylabel('monthly average temperature(℃)',color='r') #畫出y軸標籤
ax2.tick_params('y', colors='r') #畫出y軸顏色

tick = fig1.set_xticks([60,120,180,240,300,360]) #設置刻度
labels = fig1.set_xticklabels([2,4,6,8,10,12], rotation=0, fontsize =8) #設置刻度值

#創立標題
ax1.set_title("Monthly Precipitation 2565.5mm", {"size":10,"color":"blue"},loc="left")
ax2.set_title("Average annual temperature 22.7℃", {"size":10,"color":"red"},loc="right")
plt.show() #印出圖
```

```python
#problem 2 -(a) Unit_Cost 與 Unit_Price 之間關係的分佈圖
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #回歸線

fp = pd.read_csv("salse.csv")

#繪製分佈圖
sns.regplot(x="Unit_Cost", y="Unit_Price", data=fp)
sns.scatterplot(x="Unit_Cost", y="Unit_Price", hue="Product_Category", data=fp)

#設定軸標籤
plt.xlabel("Unit Cost")
plt.ylabel("Unit Price")


plt.legend(loc='upper left')
#顯示圖形
plt.show()
```

```python
#problem 2 -(b) 長條圖
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fp = pd.read_csv("salse.csv")

#先利用groupby 將資料用 country跟product_category 分類
#再特別指定revenue這個column
#再用agg("sum") 呼叫sum()算平均
group = fp.groupby(["Product_Category", "Country"])["Revenue"].agg("sum") 

fn = group.unstack() #把第二列Product_Category 轉成 行

#取得要畫圖的國家
categories = ["Australia", "Canada", "France" ,"Germany", "United Kingdom", "United States"]
#繪製長條圖
ax = fn[categories].plot(kind='bar', color=['darkblue', 'orange', 'green', 'red' ,'purple' ,'brown'])

# #設定x軸標籤
ax.set_xlabel('')
#設定y軸標籤
ax.set_ylabel('Revenue')

#顯示圖例
ax.legend(loc='upper right')

#顯示圖形
plt.show()
```
