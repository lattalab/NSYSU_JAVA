```python
#problem 1 -(a) 
import pandas as pd
import matplotlib.pyplot as plt

fp = pd.read_csv("466900-2022.csv")
monthday = [31,28,31,30,31,30,31,31,30,31,30,31] #
water = [] #
s =0
index =0
for i in monthday:
    for j in range(i): #
        s += fp["(mm)"][index]
        index += 1
    water.append(s) #
    s = 0
    
month = [i for i in range(1,13)]
plt.pie(water,  #
        labels = month, #()
        radius=1.2, #
        autopct='%.1f%%', #
         pctdistance=0.7, #
        textprops={'size':14}, #
        wedgeprops={'linewidth':3,'edgecolor':'w','width':1} #
       ) 
plt.show()
```

```python
#problem 1 -(c) 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fp = pd.read_csv("466900-2022.csv")

month = [i for i in range(1,13)]
monthday = [31,28,31,30,31,30,31,31,30,31,30,31] #
water = [] #
tem = [] #
s =0
s2 =0
index =0
for i in monthday:
    for j in range(i): #
        s += fp["(mm)"][index]
        s2 += fp["()"][index]
        index += 1
    water.append(s) #
    tem.append(s2/i) # 
    s = 0
    s2 = 0

fig, ax1 = plt.subplots()
ax1.set_xlabel('Month')

ax1.bar(month,water,color ='b')
ax1.set_ylabel('tmonthly precipitation(mm)',color='b') #y
ax1.tick_params('y', colors='b') #y

ax2 = ax1.twinx() #x y

ax2.plot(month,tem,'r-')
ax2.set_ylabel('monthly average temperature()',color='r') #y
ax2.tick_params('y', colors='r') #y

tick = fig1.set_xticks([60,120,180,240,300,360]) #
labels = fig1.set_xticklabels([2,4,6,8,10,12], rotation=0, fontsize =8) #

#
ax1.set_title("Monthly Precipitation 2565.5mm", {"size":10,"color":"blue"},loc="left")
ax2.set_title("Average annual temperature 22.7", {"size":10,"color":"red"},loc="right")
plt.show() #
```

```python
#problem 2 -(a) Unit_Cost  Unit_Price 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #

fp = pd.read_csv("salse.csv")

#
sns.regplot(x="Unit_Cost", y="Unit_Price", data=fp)
sns.scatterplot(x="Unit_Cost", y="Unit_Price", hue="Product_Category", data=fp)

#
plt.xlabel("Unit Cost")
plt.ylabel("Unit Price")


plt.legend(loc='upper left')
#
plt.show()
```

```python
#problem 2 -(b) 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fp = pd.read_csv("salse.csv")

#groupby  countryproduct_category 
#revenuecolumn
#agg("sum") sum()
group = fp.groupby(["Product_Category", "Country"])["Revenue"].agg("sum") 

fn = group.unstack() #Product_Category  

#
categories = ["Australia", "Canada", "France" ,"Germany", "United Kingdom", "United States"]
#
ax = fn[categories].plot(kind='bar', color=['darkblue', 'orange', 'green', 'red' ,'purple' ,'brown'])

# #x
ax.set_xlabel('')
#y
ax.set_ylabel('Revenue')

#
ax.legend(loc='upper right')

#
plt.show()
```
