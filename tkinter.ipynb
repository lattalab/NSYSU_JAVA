```python
# EP 5 XY隨機座標生成器
import tkinter as tk
import random

# TK視窗參數設定
win = tk.Tk() # 建立主視窗 
win.title("Random X,Y Generator")
# 大小
win.geometry("400x320+800+500") # 寬x高
# +800+500: (0,0) -> (800,500)
win.config(bg="#323232") # 背景顏色

# 文字、文字顏色、背景顏色
title_text = tk.Label(text="Random X,Y Generator",fg="skyblue",bg="#323232") 
# font = "字體 大小"
title_text.config(font="微軟正黑體 15")
title_text.pack() #封裝 

min_range = tk.Label(text="Min range", fg="white", bg="#323232")
min_range.pack()
min_entry = tk.Entry()
min_entry.pack()

max_range = tk.Label(text="Max range", fg="white", bg="#323232")
max_range.pack()
max_entry = tk.Entry()
max_entry.pack()

x_show = tk.Label(text="", fg="white", bg="#323232")
x_show.pack()
y_show = tk.Label(text="", fg="white", bg="#323232")
y_show.pack()


def Gen():
    min_value = int(min_entry.get()) # 得到輸入值，轉成int
    max_value = int(max_entry.get()) # 得到輸入值，轉成int
    x = str(random.randint(min_value,max_value)) #產生min~max之間的值
    y = str(random.randint(min_value,max_value)) #產生min~max之間的值
    x_show.config(text= "X = "+ x)
    y_show.config(text= "Y = "+ y)

generate_button = tk.Button(text="Generate",command=Gen) #可呼叫Gen()的Button
generate_button.pack()

win.mainloop()
```

```python
# EP 6 Grid 網格布局
import tkinter as tk

# TK視窗參數設定
win = tk.Tk() # 建立主視窗 
win.title("Grid")
# 大小
win.geometry("+800+500") # 寬x高
# +800+500: (0,0) -> (800,500)
win.config(bg="#323232") # 背景顏色

# Grid 網格布局
# (row,col):
# (0,0)|(0,1)
# -----|-----
# (1,0)|(1,1)

user = tk.Label(text="User ")
user.grid(row=0,column=0)

password = tk.Label(text="Password ")
password.grid(row=1,column=0)

user_entry = tk.Entry(bg = "white")
user_entry.grid(row =0,column=1)

password_entry = tk.Entry(bg = "white")
password_entry.grid(row =1,column=1)


test_label = tk.Label(text="Test")
test_label.grid(row=2,column=0)
test = tk.Entry(bg="white",font="微軟正黑體 30")
test.grid(row=2,column=1,rowspan=2) #跨兩列

win.mainloop()
```

```python
# EP 7 Pack、Place布局
import tkinter as tk

# TK視窗參數設定
win = tk.Tk() # 建立主視窗 
win.title("Pack&Place")
# 大小
win.geometry("200x200+800+500") # 寬x高
# +800+500: (0,0) -> (800,500)
win.config(bg="#323232") # 背景顏色

# Place 布局 = x,y座標布局
# anchor = 原點位置
btn = tk.Button(text="Button")
btn.place(x=20,y=20)
btn2 = tk.Button(text="Button")
# 可利用width & height 設定高度
btn2.place(anchor="center",x=100,y=100,width=100,height=50) # 絕對位置

win.mainloop()
```
