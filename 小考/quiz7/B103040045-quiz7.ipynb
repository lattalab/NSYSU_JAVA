```python
import pandas as pd

f1 = pd.read_csv("AQI-2023-01.csv")
f2 = pd.read_csv("AQI-2023-02.csv")
f3 = pd.read_csv("AQI-2023-03.csv")

# 將三個 DataFrame 結合為一個
df_all = pd.concat([f1, f2, f3])

# 找高雄市
# .isin => 篩選某列(column)中的數據是否有list中的元素
df_kh = df_all[df_all['"county"'].isin(['高雄市'])]

# 把具有空值(null)的資料整列(row)去除
df_kh = df_kh.dropna(subset=['"aqi"', '"pm2.5_avg"', '"pm10_avg"', '"so2_avg"'])

# # 將 datacreationdate 欄位轉換為日期格式
df_kh['"datacreationdate"'] = pd.to_datetime(df_kh['"datacreationdate"'])

# 將 datacreationdate 欄位的時間部分移除
df_kh['"datacreationdate"'] = df_kh['"datacreationdate"'].dt.to_period('D').dt.strftime("%Y-%m-%d")

# 計算不同監測站當天各項數值的平均
df_kh = df_kh.groupby(['"siteid"', '"sitename"', '"datacreationdate"'])[['"aqi"', '"pm2.5_avg"', '"pm10_avg"', '"so2_avg"']].mean()
df_kh = df_kh.round(2)
df_kh = df_kh.reset_index()

# 調整欄位順序並改變日期格式
df_avg = df_kh[['"siteid"', '"sitename"', '"aqi"', '"pm2.5_avg"', '"pm10_avg"', '"so2_avg"', '"datacreationdate"']]
df_avg.set_index('"siteid"',inplace=True)

# 寫進csv檔
df_avg.to_csv('AQI.csv')
```
