```python
import numpy as np
n = input("輸入你的對數:")
try:
    n = int(n)
except TypeError: # unable to change int
    n = input("輸入你的對數:")

if n >= 9:
    raise("When n>=9, it will may encounter error.")

D = dict()
for i in range(2,n+1):
    total = 10 ** i
    x = np.random.uniform(0,1,total)
    y = np.random.uniform(0,1,total)
    dis = np.sqrt((x-0.5) ** 2 + (y-0.5) ** 2)
    cond = dis <= 0.5
    pi = 4 * len(dis[cond])/ total
    D[i] = round(pi,7)

print("PI dictionary (2 - {}):".format(n))
print(D)
```

```python
# 加分題 圖形化
from matplotlib import pyplot as plt
xticks = ["10^" + str(i) for i in range(2,n+1)]
fig, ax = plt.subplots(figsize = (16, 5))

ax.set_title('Monte Carlo', fontsize=20)
ax.plot(xticks, list(D.values()), marker = "o") 
plt.axhline(y=3.14159265, ls="--", lw=2) # 畫pi的漸進線

# 設置X軸、Y軸標籤
ax.set_xlabel('Node')   
ax.set_ylabel('Value of Pi')

# 在每個點標註y值
for i, j in enumerate(list(D.values())):
    ax.text(xticks[i], j, str(j), ha='left', va='bottom')
    
ax.text(xticks[0],3.14,'True value of Pi', ha='left', va='bottom') 
plt.show()
```
