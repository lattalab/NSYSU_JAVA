```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog

#初始化tkinter
#-----------------------------------------------------考試填空部分------------------------------------------------------
window = tk.Tk()
window.title("日本確診人數")
window.geometry("500x450")
text = tk.Text(window)
#-----------------------------------------------------考試填空部分------------------------------------------------------
# file_path創建存放檔案的路徑
file_path = tk.StringVar()
df = None
new_df=None

def open_file():
    global df
    # 打開對話框，選擇 CSV 文件
    #-----------------------------------------------------考試填空部分-----------------------------------------------------
    # 創建變量
    global file_path

    #刪除Text元件的文字
    text.delete(1.0, tk.END)

    # 打開對話框，選擇 CSV 文件
    file_path.set(filedialog.askopenfilename(filetypes=[("CSV Files", "*.csv")]))
    if file_path.get() != "":
        df = pd.read_csv(file_path.get())
        text.insert(tk.END, df.head(15))
    #-----------------------------------------------------考試填空部分------------------------------------------------------

def fill_nans():
    print()
    #用0填補df空缺值
    #-----------------------------------------------------考試填空部分------------------------------------------------------
    global df
    df['new_cases'].fillna(0.0, inplace=True)
    df['new_deaths'].fillna(0.0, inplace=True)

    #刪除Text元件的文字
    text.delete(1.0, tk.END)
    text.insert(tk.END, df.head(15))
    #-----------------------------------------------------考試填空部分------------------------------------------------------

def delete_column():
    global new_df
    #刪除不能計算平均值的欄位(不要更改原始df，新建一個new_df)
    #-----------------------------------------------------考試填空部分------------------------------------------------------
    new_df = df.copy()
    del new_df["location"]
    del new_df["date"]
    

    #刪除Text元件的文字
    text.delete(1.0, tk.END)
    text.insert(tk.END, new_df.head(15))
    #-----------------------------------------------------考試填空部分------------------------------------------------------

def calculate_means():
    #計算new_df各欄位的平均值
    #-----------------------------------------------------考試填空部分------------------------------------------------------
    global new_df

    a =pd.Series([new_df["new_cases"].mean(),new_df["new_deaths"].mean(),new_df["population_density"].mean()]
    ,index = ["new_cases","new_deaths","population_density"])
    
    #刪除Text元件的文字
    text.delete(1.0, tk.END)

    text.insert(tk.END, a)
    #-----------------------------------------------------考試填空部分------------------------------------------------------

def calculate_sums():
    # 利用df計算各月份的確診人數(new_cases)、死亡人數總和(new_deaths)
    #-----------------------------------------------------考試填空部分------------------------------------------------------
    global df
    df["date"] = pd.to_datetime(df["date"])
    df["date"] = df["date"].dt.to_period('M')
    df  = df.groupby(["date"])[["new_cases","new_deaths"]].sum()
    #刪除Text元件的文字
    text.delete(1.0, tk.END)

    text.insert(tk.END, df)
    #-----------------------------------------------------考試填空部分------------------------------------------------------


# 創建Button和Text
#-----------------------------------------------------考試填空部分------------------------------------------------------
# button
button1 = tk.Button(text="打開CSV文件") # create a button object
button1.config(command=open_file)

button2 = tk.Button(text="填補缺失值") # create a button object
button2.config(command=fill_nans)

button3 = tk.Button(text="刪除欄位") # create a button object
button3.config(command=delete_column)

button4 = tk.Button(text="計算平均值") # create a button object
button4.config(command=calculate_means)

button5 = tk.Button(text="各月份的確診人數、死亡人數總和") # create a button object
button5.config(command=calculate_sums)
#-----------------------------------------------------考試填空部分------------------------------------------------------

button1.pack(pady=10)
button2.pack(pady=10)
button3.pack(pady=10)
button4.pack(pady=10)
button5.pack(pady=10)
text.pack(padx=10, pady=10, fill="both", expand=True)


# 運行應用程序
#-----------------------------------------------------考試填空部分------------------------------------------------------
window.mainloop()
#-----------------------------------------------------考試填空部分------------------------------------------------------
```
